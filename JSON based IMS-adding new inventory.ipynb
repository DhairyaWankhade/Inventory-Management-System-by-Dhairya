## Read Data from json

```python
import json
fd = open("record.json",'r')
r = fd.read()
fd.close()
```

```python
record = json.loads(r)
```

## To Add New Item into Inventory

```python
book_id = input("Enter Book ID: ")
book_name = input("Enter Book Name: ")
author = input("Enter Author Name: ")
publisher = input("Enter Publisher Name: ")
edition = input("Enter Edition: ")
price = input("Enter Price of Book: ")
quant = int(input("Enter quantity of Books: "))

for key in record.keys():
    if key == book_id:
        quant+=record[book_id]['quantity']
record[book_id] = {"name": book_name, "Author": author, "publisher": publisher, "edition": edition, "price": price, "quantity": quant}

js = json.dumps(record)

fd = open("record.json",'w')
fd.write(js)
fd.close()
               
```
