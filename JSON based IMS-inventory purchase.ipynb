## Read from inventory

```python
import json
fd = open("record.json",'r')
r = fd.read()
fd.close()
records = json.loads(r)
```

```python
records
```

## Purchase

```python
ui_prod  = str(input("Enter the product Id: "))
ui_quant = int(input("Enter the quantity: "))

for key in records.keys():
    if (key == ui_prod):
        if(records[ui_prod]['quantity']>=ui_quant):
            print("Your Bill ")
            print("******************************")
            print("Product / Book: ", records[ui_prod]['name'])
            print("Price: ", records[ui_prod]['price'])
            print("Quantity: ", ui_quant)
            print("-------------------------------")
            print("Billing Amount: ", int(records[ui_prod]['price']) * int(ui_quant))
            print("******************************")
            
            records[ui_prod]['quantity'] = int(records[ui_prod]['quantity']) - int(ui_quant)
            
        else:
            print("\nOnly "+ str(records[ui_prod]['quantity']) + " items of this product is available.")
            ui_item = bool(int(input("\nDo you want to buy "+ str(records[ui_prod]['quantity']) + " items?\nEnter 1 for 'YES' or 0 for 'NO' : ")))
            if(ui_item == True):
                print("\nYour Bill ")
                print("******************************")
                print("Product / Book: ", records[ui_prod]['name'])
                print("Price: ", records[ui_prod]['price'])
                print("Quantity: ",  records[ui_prod]['quantity'])
                print("-------------------------------")
                print("Billing Amount: ", records[ui_prod]['price'] * records[ui_prod]['quantity'])
                print("******************************")
                ui_quant = records[ui_prod]['quantity']
                records[ui_prod]['quantity'] = records[ui_prod]['quantity'] - records[ui_prod]['quantity']
                
            else:
                print("\nOk, You can buy other products.\nEnjoy your shopping.")
```

## Update inventory after purchase

```python
js = json.dumps(records)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```
